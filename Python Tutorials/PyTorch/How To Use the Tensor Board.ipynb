# How To Use the Tensor Board

In this video, we will learn to use the tensor board to visualize and analyze our model/training pipeline. Tensor board is a visualization tool kit to experiment with our models. It is made by the TensorFlow individuals, but can be used with `pytorch`. More information can be found here: https://www.tensorflow.org/tensorboard.

The tensor board can be used for: 

* Tracking and visualizing metrics such as loss and accuracy
* Visualizing the model graph (ops and layers)
* Viewing histograms of weights, biases, or other tensors as they change over time
* Projecting embeddings to a lower dimensional space
* Displaying images, text, and audio data
* Profiling TensorFlow programs
* And much more

So, how can we use this in our code? We will use the code from tutorial number 13, which was a feed forward NN. As a brief recape, in tutorial number 13, we:

* Used MNIST data for digit classification
* Plotted some of the images
* Created a feed forward NN with 1 linear layer then ReLU activation, then another linear layer
* Then we set up the training pipeline with loss and optimizer
* We did training: forward pass, backward pass, and weights update
* Finally, we used tested the model and determine the accuracy

Now, let's move on to using the tensor board. Prior to using the tensor board, we can install it with `pip install tensorboard`. We did not need to install the whole `tensorflow` library, just `tensorboard`. To start `tensorboard`, in the command prompt/terminal, we write: `tensorboard --logdir=runs`. If you run into any issues with this and are using Jupyter Notebook, you can use (if you have no dashboards active or written any data, nothing will really show):
