# Transfer Learning

In this tutorial, we will talk about transfer learning and how it can be implemented in `pytorch`. Transfer learning is a machine learning method where a model developed on a first task is reused as the starting point for a model on a second task. For example, we can train a model to identify birds and cats and then use the same model, with the last layer modified, to classify bees and dogs. This allows for rapid generation of new models. This is important because the training of a new model can be very time consuming. If we use a pre-trained model, then we typically exchange only the last layer and we do not need to train the whole model again. Transfer learning can have pretty good performance results.



Now, let's look at a concrete example in `pytorch`. In this example, we are using the pre-trained ResNet-18 CNN. This is a CNN has been trained on more than a million images from the image net database. This network is 18 layers deep and can classify images into 1,000 object categories. Now, in our example, we only have 2 classes (bees and ants).
